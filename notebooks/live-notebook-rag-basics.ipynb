
# Addressig of the basics of Embeddings

```python
text1 = "Lucas is an AI engineer who loves to talk about AI augmented systems"

text2 = "John is a software engineer who loves to talk about software development"

text3_different = "The weather in Tokyo is sunny"
```

```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings()

embedding1 = embeddings.embed_query(text1)
embedding2 = embeddings.embed_query(text2)
embedding3 = embeddings.embed_query(text3_different)
```

```python
def cosine_similarity(vectors):
    # Initialize empty similarity matrix
    n = len(vectors)
    similarity_matrix = np.zeros((n, n))
    
    # Calculate cosine similarity between all pairs
    for i in range(n):
        for j in range(n):
            # Get dot product
            dot_product = np.dot(vectors[i], vectors[j])
            
            # Get magnitudes
            norm_i = np.sqrt(np.sum(vectors[i]**2))
            norm_j = np.sqrt(np.sum(vectors[j]**2))
            
            # Calculate cosine similarity
            similarity_matrix[i][j] = dot_product / (norm_i * norm_j)
            
    return similarity_matrix

import numpy as np

# Reshape the embeddings for sklearn's cosine_similarity
embeddings_array = np.array([embedding1, embedding2, embedding3])

# Calculate pairwise similarities
similarity_matrix = cosine_similarity(embeddings_array)

print("Similarity scores:")
print(f"text1 vs text2: {similarity_matrix[0][1]:.3f}")
print(f"text1 vs text3: {similarity_matrix[0][2]:.3f}") 
print(f"text2 vs text3: {similarity_matrix[1][2]:.3f}")
```
